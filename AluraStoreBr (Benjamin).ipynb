### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
loja_faturamento = (loja['Preço']+loja['Frete']).sum()
print(loja_faturamento)
```

```python
loja_faturamento2 = (loja2['Preço']+loja2['Frete']).sum()
print(loja_faturamento2)
```

```python
loja_faturamento3 = (loja3['Preço']+loja3['Frete']).sum()
print(loja_faturamento3)
```

```python
loja_faturamento4 = (loja4['Preço']+loja3['Frete']).sum()
print(loja_faturamento4)
```

# 2. Vendas por Categoria


```python
Categoria_Loja = loja.groupby('Categoria do Produto')['Preço'].sum()
print(Categoria_Loja)
```

```python
Categoria_Loja2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
print(Categoria_Loja2)
```

```python
Categoria_Loja3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
print(Categoria_Loja3)
```

```python
Categoria_Loja4 = loja4.groupby('Categoria do Produto')['Preço'].sum()
print(Categoria_Loja4)
```

# 3. Média de Avaliação das Lojas

```python
media_Avaliação = loja['Avaliação da compra'].mean()
print("Avaliação media da loja 1: " + str(media_Avaliação))
```

```python
media_Avaliação3 = loja3['Avaliação da compra'].mean()
print("Avaliação media da loja 3: " + str(media_Avaliação3))
```

```python
media_Avaliação4 = loja4['Avaliação da compra'].mean()
print("Avaliação media da loja 4: " + str(media_Avaliação4))
```

# 4. Produtos Mais e Menos Vendidos

```python
mais_vendido = loja['Produto'].value_counts().idxmax()
menos_vendido = loja['Produto'].value_counts().idxmin()
print("produto mais vendido 1: "+mais_vendido + '\n'+"produto menos vendido 1: "+menos_vendido)
```

```python
mais_vendido2 = loja2['Produto'].value_counts().idxmax()
menos_vendido2 = loja2['Produto'].value_counts().idxmin()
print("produto mais vendido 2: "+mais_vendido2 + '\n'+"produto menos vendido 2: "+menos_vendido2)
```

```python
mais_vendido3 = loja3['Produto'].value_counts().idxmax()
menos_vendido3 = loja3['Produto'].value_counts().idxmin()
print("produto mais vendido 3: "+mais_vendido3 + '\n'+"produto menos vendido 3: "+menos_vendido3)
```

```python
mais_vendido4 = loja4['Produto'].value_counts().idxmax()
menos_vendido4 = loja4['Produto'].value_counts().idxmin()
print("produto mais vendido 4: "+mais_vendido4 + '\n'+"produto menos vendido 4: "+menos_vendido4)
```

# 5. Frete Médio por Loja

# 📊 Relatório de Análise das Lojas – Alura Store

## Objetivo
Identificar qual das quatro lojas apresenta o menor desempenho para recomendação de venda ao Sr. João.

## Métricas analisadas
- **Faturamento total**  
- **Média de avaliação dos clientes**  
- **Produto mais vendido e menos vendido**  
- **Custo médio do frete**

## Resultados

- A **Loja 4** apresenta o **menor faturamento**, ficando atrás das demais.  
- A diferença de **avaliação média** entre as lojas é pequena, sem grandes destaques.  
- O **frete médio** da Loja 4 é o menor, porém este fator não compensa a baixa receita.  
- Produtos mais e menos vendidos variam entre as lojas e não alteram o cenário geral.

## Conclusão
Com base nas métricas, a **Loja 4** apresenta o **pior desempenho financeiro**.  
**Recomendação:** vender a **Loja 4** para gerar capital e investir em novos negócios.
